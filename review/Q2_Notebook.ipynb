## Visualizing Data & Table Methods


This is an example of how to make a table from scratch

```python
# Make the neuro table in datascience package
Month = make_array(1,1,2)
Journal = make_array("Journal of Neuroscience", "Neuron", "Nature Neuroscience")
Words = make_array(6780, 7980, 5500)
Citations = make_array(568,200,870)
Method = make_array('MRI','EEG','Optogenetics')
Species = make_array('Human', 'Human', 'Animal')

neuro = Table().with_columns('Month', Month, 'Journal', Journal, 'Words', Words, 'Citations', Citations, 'Method', Method, 'Species', Species)

neuro
```

## Another way to make tables that we recently learned: `.with_rows`

```python
#@title Initialize an Empty table with Column Names
# Empty table with columns and no rows
brains = Table(['Month', 'Journal', 'Words', 'Citations', 'Method', 'Species'])
brains
```

```python
#@title Make a table with rows
brains = brains.with_rows([
    [1, 'Journal of Neuroscience', 6780, 568,'MRI', 'Human'],
    [1, 'Neuron', 7980, 200, 'EEG', 'Human'],
    [2, 'Nature Neuroscience', 5500, 870,'Optogenetics', 'Animal'],
])
brains
```
